## Teste de hipótese

**H<sup>null</sup> - hipótese nula:**
**A distribuição da taxa da gorjeta é a mesma nos dois grupos**

**H<sup>alt</sup> - hipótese alternativa:**
**A distribuição da taxa da gorjeta não é a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'Não'").porcentagem
```

```python
sem_sobremesa
```

```python
ranksums(sobremesa, sem_sobremesa)
```

**Apesar de os dois grupos terem dado gorjetas de valores diferentes entre si, a diferença é insignificante do ponto de vista matemático. Ou seja, não poderemos dizer que a população geral será diferente também.
Cientes disso, continuaremos somente com nossa hipótese nula, não aceitando mais a hipótese alternativa. Para aceitarmos esta última, o valor de pvalue deveria ser igual ou menor do que 0.05. Como nosso resultado foi maior do que isso, poderemos descartar esta hipótese alternativa.**

```python
r = ranksums(sobremesa, sem_sobremesa)
print('O valor do p-value é {}'.format(r.pvalue))
```

**Por tanto, a distribuição da taxa da gorjeta é a mesma nos dois grupos**

## Teste de hipótese

**H<sup>null</sup> - hipótese nula:**
**A distribuição do valor da conta é igual no sábado e no domingo**

**H<sup>alt</sup> - hipótese alternativa:**
**A distribuição do valor da conta não é igual no sábado e no domingo**

```python
valor_contas_domingo = gorjetas.query("dia_da_semana == 'Domingo'").valor_da_conta
valor_contas_domingo
```

```python
valor_contas_sabado = gorjetas.query("dia_da_semana == 'Sábado'").valor_da_conta
valor_contas_sabado
```

```python
r2 = ranksums(valor_contas_domingo, valor_contas_sabado)
print(f'O valor do p-value é igual a {r2.pvalue}')
```

**Por tanto, a distribuição do valor da conta é igual no sábado e no domingo**
